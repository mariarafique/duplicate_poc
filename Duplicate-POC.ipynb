```python
import numpy as np 
import pandas as pd 
import os
from PIL import Image
import imagehash
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, f1_score, precision_score, accuracy_score, balanced_accuracy_score
```

```python
def assignLabel(imageName):
    if 'duplicate' in imageName:
        return 1
    return 0
    
def createDataFrame(folderPath):
    dataFrame = pd.DataFrame([], columns=['image name','hash', 'label'])
    data = []
    for filename in os.listdir(folderPath):
        imageName=os.path.join(folderPath,filename)
        print(imageName)
        imageHash = {}
        original_hash = imagehash.crop_resistant_hash(Image.open(imageName)) #imagehash.phash(Image.open('f31.png'))
        imageHash['image name'] = filename
        imageHash['hash'] = str(original_hash)
        imageHash['label'] = assignLabel(imageName)


        data.append(imageHash)
    return pd.DataFrame(data, columns=['image name','hash', 'label'])
```

```python
imagesDataFrame =createDataFrame('dataset')
```

```python
X = imagesDataFrame.drop(columns=["hash", "image name"])
y = imagesDataFrame["label"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=123)

knn = KNeighborsClassifier(n_neighbors=3,metric='hamming')
knn.fit(X, y)
```

```python
predicted_labels=knn.predict(X_test)# test
```

```python
print(accuracy_score(y_test, predicted_labels), "accuracy score")
print(balanced_accuracy_score(y_test, predicted_labels), "accuracy score")
```
